### Ćwiczenie 1

```python
#funkcja range jest zamknięta z prawej strony, dlatego z+1
def check_range(x,y,z):
  if x in range(y,z+1): return f'{x} znajduje się pomiędzy {y} a {z}'
  return f'{x} nie znajduje się pomiędzy {y} a {z}'
```

```python
check_range(34, 9, 228)
```

```python
check_range(7, 2, 5)
```

```python
#funkcja range jest zamknięta z prawej strony, dlatego z+1
def bool_range(x,y,z):
  return x in range(y,z+1)
```

```python
bool_range(7, 5, 20)
```

```python
bool_range(67, 22, 25)
```

### Ćwiczenie 2


```python
def unique_list(list):
  unique_items = []
  for item in list:
    if item not in unique_items:
      unique_items.append(item)
  return unique_items
```

```python
my_list = [1,3,5,6,4,3,2,3,3,4,3,4,5,6,6,4,3,2,12,3,5,63,4,5,3,3,2]
unique_list(my_list)
```

```python
#inny sposób wykonania tej operacji bez funkcji
unique_items = []
for item in my_list:
  if item not in unique_items:
    unique_items.append(item)
print(unique_items)

#lub
print(list(set(my_list)))
```

### Ćwiczenie 3

```python
def sphere_volume(radius):
  #4/3 * pi * r^3
  return 4/3 * 3.14 * radius**3
```

```python
sphere_volume(2)
```

### Ćwiczenie 4

```python
def num_fact(number):
  if number < 0: return "ERROR"
  if number == 0: return 1
  return number * num_fact(number-1)
```

```python
num_fact(12)
```
